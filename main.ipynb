```python
from sklearn.datasets import load_wine
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
wine_data = load_wine()
wine_data
```

```python
df = pd.DataFrame(data=wine_data.data, columns=wine_data.feature_names)
df['target'] = wine_data.target

df.describe()
```

```python
df.columns
```

```python
df.info()
```

```python
df['target']
```

```python
df['target'] = df['target'].map({0: 'class_0', 1: 'class_1', 2: 'class_2'})
```

```python
df.head()
```

```python
small_df = df[["alcohol", "malic_acid", "target"]]
small_df.head()
small_df['alcohol'] += 10
small_df.head()
```

```python
first_5_rows = small_df.iloc[:5, :2].values
first_5_rows
```

```python
mean_first_5_rows = np.mean(first_5_rows)
mean_first_5_rows
```

```python
df.head()
```

```python
df.rename(columns={'target': 'class'}, inplace=True)
df
```

```python
avg_alcohol = df['alcohol'].mean()
avg_alcohol
```

```python
heavy_alcohol = df[df['alcohol'] > avg_alcohol]
heavy_alcohol.head(10)
```

```python
df.columns
```

```python
plt.figure(figsize=(8, 6))

for wine_class in df['class'].unique():
    class_data = df[df['class'] == wine_class]
    plt.scatter(class_data['alcohol'], class_data['malic_acid'], label=wine_class) 

plt.xlabel('Alcohol')   
plt.ylabel('Malic Acid')
plt.title('Alcohol and Malic Acid content in wine')

plt.legend(title="Wine Classes")
plt.show()
```

```python
for wine_class in df['class'].unique():
    class_data = df[df['class'] == wine_class]
    plt.hist(class_data['alcohol'], alpha=0.5, label=wine_class, bins=10)

plt.xlabel('Alcohol')
plt.ylabel('Frequency')
plt.title('Alcohol content in wine')
plt.legend(title="Wine Classes")
plt.show()
```
